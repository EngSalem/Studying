## implement kmeans

```python
import numpy as np
import random 
```

```python
class kmeans:
    def __init__(self, k, max_iters=100):
        self.k = k
        self.max_iters = max_iters
        self.centroids = None
    
    def _get_distance(self, x, centroids):
        distances = []
        for centroid in centroids:
            dist = np.sqrt(np.sum((x-centroid)**2))
            distances.append(dist)
        return distances

    def fit(self, X):
        '''
        X = N x D data matrix
        '''
        N, D = X.shape

        # random init of centroids
        self.centroids = X[np.random.choice(N, self.k, replace=False)]

        for _ in range(self.max_iters):
            # assign each point to closest centroid
            clusters = [[] for _ in range(self.k)]

            for x in X:
                distances = self._get_distance(x, self.centroids)
                closest_centroid = np.argmin(distances)
                clusters[closest_centroid].append(x)

            # update centroids 
            new_centroids = []
            for i in range(self.k):
                if len(clusters[i]) > 0:
                    # Calculate mean of points in cluster
                    new_centroids.append(np.mean(clusters[i], axis=0))
                else:
                    # Keep old centroid if no points assigned
                    new_centroids.append(self.centroids[i])
            
            new_centroids = np.array(new_centroids)

            # Check for convergence
            if np.allclose(self.centroids, new_centroids):
                break
                
            self.centroids = new_centroids

    def predict(self, X):
        '''
        X = N x D data matrix
        '''
        if self.centroids is None:
            raise ValueError("Model must be fitted before making predictions")
            
        predictions = []
        for x in X:
            distances = self._get_distance(x, self.centroids)
            closest_centroid = np.argmin(distances)
            predictions.append(closest_centroid)
        return np.array(predictions)
```

```python
# Test the K-means implementation
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, 
                       random_state=0)

# Apply our K-means
kmeans_model = kmeans(k=4, max_iters=100)
kmeans_model.fit(X)
predictions = kmeans_model.predict(X)

# Plot results
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis')
plt.title('True Clusters')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=predictions, cmap='viridis')
plt.scatter(kmeans_model.centroids[:, 0], kmeans_model.centroids[:, 1], 
           c='red', marker='x', s=100, label='Centroids')
plt.title('K-Means Results')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Final centroids:\n{kmeans_model.centroids}")
```
