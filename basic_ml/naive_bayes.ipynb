## implementing naive bayes
## key idea , during fit
## 1. Learn prior (frequency of labels/total labels)
## 2. Learn feature relvance 
## during prediction apply laplace smoothing 

```python
import numpy as np

class NaiveBayes:
    def __init__(self):
        self.class_priors = {}
        self.feature_likelihoods = {}

    def fit(self, X,y):
        # Compute class priors
        for cls in set(y):
            self.class_priors[cls] = np.mean(y==cls)

        # compute feature likelihoods
        for cls in set(y):
            self.feature_likelihoods[cls] = np.mean(X[y==cls], axis=0)
    
    def predict(self,X):
        predictions = []
        for x in X:
            class_scores= {}

            for cls in self.class_priors:
                # Start with log prior
                class_scores[cls] = np.log(self.class_priors[cls])

                # Add log likelihoods
                class_scores[cls] += np.sum(x * np.log(self.feature_likelihoods[cls]) + (1-x)*np.log(1-self.feature_likelihoods[cls]))

            # Choose the class with the highest score
            predictions.append(max(class_scores, key=class_scores.get))
        return np.array(predictions)
```
