## implementing naive bayes
## key idea , during fit
## 1. Learn prior (frequency of labels/total labels)
## 2. Learn feature relvance 
## during prediction apply laplace smoothing 

```python
import numpy as np
class NaiveBayes:
    def __init__(self):
        self.class_priors = {}
        self.feature_likelihoods = {}
    
    def fit(self, X, y):
        # compute class priors
        total_samples = len(y)
        for cls in set(y):
            self.class_priors[cls] = np.sum(y==cls)/total_samples 
        
        for cls in self.class_priors:
            cls_samples = X[y==cls]
            self.feature_likelihoods[cls] = np.mean(cls_samples, axis=0)
    def predict(self, X):
        predictions = []
        for x in X:
            class_scores = {}
            for cls in self.class_priors:
                # compute log likelihood
                likelihood = np.prod(self.feature_likelihoods[cls]**x * (1-self.feature_likelihoods[cls])**(1-x))
                class_scores[cls] = np.log(self.class_priors[cls]) + np.log(likelihood)
            predictions.append(max(class_scores, key=class_scores.get))
        return np.array(predictions)
```
